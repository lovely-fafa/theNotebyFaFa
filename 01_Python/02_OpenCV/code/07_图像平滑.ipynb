```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
dogsp = cv.imread('./image/dogsp.jpeg')
doggau = cv.imread('./image/dogGauss.jpeg')
```

# 1 均值滤波

```python
dog = cv.blur(dogsp, (5, 5))
plt.imshow(dog[:, :, ::-1])
```

```python
# 2 高斯滤波
```

```python
dog = cv.GaussianBlur(doggau, (3, 3), 1)
plt.imshow(dog[:, :, ::-1])
```

# 3 中值滤波

```python
dog = cv.medianBlur(dogsp, 3)
plt.imshow(dog[:, :, ::-1])
```
