```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('./image/wulin.jpeg')
plt.imshow(img[:, :, ::-1])
```

```python
bai = cv.imread('./image/bai.jpeg')
plt.imshow(bai[:, :, ::-1])
```

# 1 

```python
res = cv.matchTemplate(img, bai, cv.TM_CCORR)
```

```python
plt.imshow(res, cmap=plt.cm.gray)
```

```python
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
```

```python
h, w = bai.shape[:2]
```

```python
bottom_right = (max_loc[0] + w, max_loc[1] + h)
```

```python
cv.rectangle(img, max_loc, bottom_right, (0, 255, 0))
```
