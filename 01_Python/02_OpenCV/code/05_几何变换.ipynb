```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
kids = cv.imread('image/kids.jpg')
```

```python
plt.imshow(kids[:, :, ::-1])
```

# 1 图像缩放

```python
# 绝对尺寸
rows, cols = kids.shape[:2]
(rows, cols)
```

```python
# 缩放
res = cv.resize(kids, (2 * rows, 2 * cols))
```

```python
plt.imshow(res[:, :, ::-1])
```

```python
# 相对缩放
res = cv.resize(kids, None, fx=0.5, fy=0.5)
plt.imshow(res[:, :, ::-1])
```

# 2 图像平移

```python
M = np.float32([
    [1, 0, 100],
    [0, 1, 50]
])
```

```python
res2 = cv.warpAffine(kids, M, (cols, rows))
plt.imshow(res2[:, :, ::-1])
```

# 3 图像旋转

```python
# 旋转矩阵
M = cv.getRotationMatrix2D((cols / 2, rows / 2), 45, 0.5)
```

```python
res3 = cv.warpAffine(kids, M, (cols, rows))
plt.imshow(res3[:, :, ::-1])
```

# 4 仿射变换

```python
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[100, 100], [200, 50], [100, 250]])
```

```python
M = cv.getAffineTransform(pts1, pts2)
M
```

```python
res4 = cv.warpAffine(kids, M, (cols, rows))
plt.imshow(res4[:, :, ::-1])
```

# 5 投射变换

```python
pts1 = np.float32([[56, 65], [368, 52], [28, 387], [389, 390]])
pts2 = np.float32([[100, 145], [300, 100], [80, 290], [310, 300]])
```

```python
M = cv.getPerspectiveTransform(pts1, pts2)
M
```

```python
res5 = cv.warpPerspective(kids, M, (cols, rows))
plt.imshow(res5[:, :, ::-1])
```
