```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
rain = cv.imread('image/rain.jpg')
```

```python
plt.imshow(rain[:, :, ::-1])
```

```python
view = cv.imread('image/view.jpg')
```

```python
plt.imshow(view[:, :, ::-1])
```

```python
img1 = cv.add(rain, view)
plt.imshow(img1[:, :, ::-1])
```

```python
img2 = rain + view
plt.imshow(img2[:, :, ::-1])
```

```python
img3 = cv.addWeighted(img1,0.7,img2,0.3,0)
plt.imshow(img3[:, :, ::-1])
```
