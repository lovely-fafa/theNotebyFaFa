```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# 1 腐蚀与膨胀

```python
img = cv.imread('./image/letter.png')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
# 核结构
kenel = np.ones((5, 5), np.uint8)
```

```python
# 腐蚀
img2 = cv.erode(img, kenel)
```

```python
plt.imshow(img2[:, :, ::-1])
```

```python
# 膨胀
img3 = cv.dilate(img, kenel)
```

```python
plt.imshow(img3[:, :, ::-1])
```

# 开闭运算

```python
open = cv.imread('./image/letteropen.png')
```

```python
plt.imshow(open[:, :, ::-1])
```

```python
kenel = np.ones((15, 15), np.uint8)
```

```python
cvopen = cv.morphologyEx(open, cv.MORPH_OPEN, kenel)
```

```python
plt.imshow(cvopen[:, :, ::-1])
```

```python
close = cv.imread('./image/letterclose.png')
plt.imshow(close[:, :, ::-1])
```

```python
cvclose = cv.morphologyEx(close, cv.MORPH_CLOSE, kenel)
plt.imshow(cvclose[:, :, ::-1])
```

# 3 黑帽和礼帽

```python
top = cv.morphologyEx(open, cv.MORPH_TOPHAT, kenel)
plt.imshow(top[:, :, ::-1])
```

```python
black = cv.morphologyEx(close, cv.MORPH_BLACKHAT, kenel)
plt.imshow(black[:, :, ::-1])
```
